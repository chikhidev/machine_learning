```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("yasserh/bmidataset")

print("Path to dataset files:", path)
```

```python
raw = pd.read_csv(f"{path}/bmi.csv")
raw
```

```python
x, y = raw['Height'].to_numpy(), raw['Weight'].to_numpy()
x = x[:20]
y = y[:20]
min_x, max_x = min(x), max(x)
min_y, max_y = min(y), max(y)
```

```python
def norm(val, min_, max_):
  return (val - min_) / (max_ - min_)
```

```python
ready_x = [norm(e, min_x, max_x) for e in x]
ready_y = [norm(e, min_y, max_y) for e in y]
```

```python
from matplotlib.lines import lineStyles
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(ready_x, ready_y, marker=".", linestyle="")
ax.set_xlim(0,1)
ax.set_ylim(0,1)
```
