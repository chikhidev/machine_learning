```python
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np
```

```python
x = np.array([[1], [2], [3], [4], [5]])
```

```python
y = np.array([45, 50, 65, 70, 80])
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y)
```

```python
to_be_predected = 0.5
while True:
  to_be_predected = input("what to predict:")
  if to_be_predected == 'n':
    break
  got = model.predict([[float(to_be_predected)]])
  print("Predicted:", got)
  print("Slot:", model.coef_)
  print("Intercept:", model.intercept_)
  x_level = np.linspace(min(x), max(x), 100).reshape(-1, 1)
  print(x_level)
  y_level = model.predict(x_level)
  plt.plot(x, y_level)
```
