```python
import numpy as np
```

```python
# Word,"Vector [Royalty, Femininity]"
# King,"[0.9,0.1]"
# Queen,"[0.9,0.9]"
# Man,"[0.1,0.1]"
# Woman,"[0.1,0.9]"
# Apple,"[0.0,0.5]"

embeddings = {
    'king': np.array([.9, .1]),
    'queen': np.array([.9, .9]),
    'man': np.array([.1, .1]),
    'women': np.array([.1, .9]),
    'apple': np.array([.0, .5])
}
```

```python
royality = embeddings['king'] - embeddings['man']
royality
```

```python
#$$King - Man + Woman = ?$$
queen = embeddings['women'] + royality
queen
```

```python
score = (queen[0] * embeddings['apple'][0]) + (queen[1] * embeddings['apple'][1])
score
```

```python
score = (queen[0] * embeddings['man'][0]) + (queen[1] * embeddings['man'][1])
score
```

```python
score = (queen[0] * embeddings['king'][0]) + (queen[1] * embeddings['king'][1])
score
```
