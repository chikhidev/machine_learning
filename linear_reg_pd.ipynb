```python
model = LinearRegression()
```

```python
file_ = pd.read_csv("/root/student_exam_scores.csv")
file_
```

```python
data = file_.to_numpy()
data
```

```python
print("data length:", len(data))
train_max = int((len(data) - 1) * .95)
data_order = ['hours_studied', 'sleep_hours', 'attendance_percent', 'previous_scores']
X = file_[data_order].to_numpy()
y = file_['exam_score'].to_numpy()
```

```python
X_train, X_test = X[:train_max], X[train_max:]
y_train, y_test = y[:train_max], y[train_max:]
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_, model.intercept_
```

```python
print(data_order)
for i in range(len(X_test)):
  got = model.predict([X_test[i]])
  print(f"test rig {X_test[i]} got: {got}, expected: {y_test[i]}, diff => {abs(y_test[i] - got)}")
```
